```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(15)
dataset=np.random.normal(100,25,200)
print(dataset)
```

```python
figure=plt.figure(figsize=(10,8))
plt.boxplot(dataset)
plt.grid()
plt.show()
```
